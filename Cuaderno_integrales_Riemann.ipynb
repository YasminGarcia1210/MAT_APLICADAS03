# Sumas de Riemann

```python
import numpy as np

def f(x):
    return x**2

def suma_riemann_derecha(func, a, b, particiones):
    delta_x = (b - a) / particiones
    x_vals = np.linspace(a, b, particiones + 1)
    suma = np.sum(func(x_vals[1:]) * delta_x)
    return suma

def suma_riemann_izquierda(func, a, b, particiones):
    delta_x = (b - a) / particiones
    x_vals = np.linspace(a, b, particiones + 1)
    suma = np.sum(func(x_vals[:-1]) * delta_x)
    return suma

def suma_riemann_punto_medio(func, a, b, particiones):
    delta_x = (b - a) / particiones
    x_vals = np.linspace(a, b, particiones + 1)
    x_midpoints = (x_vals[:-1] + x_vals[1:]) / 2
    suma = np.sum(func(x_midpoints) * delta_x)
    return suma

# Parámetros
limite_inferior = 0
limite_superior = 1
numero_particiones = 100

# Calcular las sumas de Riemann
resultado_derecha = suma_riemann_derecha(funcion_a_integrar, limite_inferior, limite_superior, numero_particiones)
resultado_izquierda = suma_riemann_izquierda(funcion_a_integrar, limite_inferior, limite_superior, numero_particiones)
resultado_punto_medio = suma_riemann_punto_medio(funcion_a_integrar, limite_inferior, limite_superior, numero_particiones)

# Mostrar resultados
print(f"Resultado Riemann Derecha: {resultado_derecha}")
print(f"Resultado Riemann Izquierda: {resultado_izquierda}")
print(f"Resultado Riemann Punto Medio: {resultado_punto_medio}")
```

# Regla de Simpson

La regla de Simpson es una técnica numérica para aproximar el valor de una integral definida. Se utiliza para estimar el área bajo la curva de una función en un intervalo específico. La regla de Simpson se basa en la idea de aproximar la función por segmentos de parábolas.

La fórmula básica de la regla de Simpson para aproximar la integral definida de una función f(x) en el intervalo [a,b] con un número par de subintervalos$(n)$

$y = \int_{a}^{b} f(x)  dx = \frac{𝚫}{n}(f(x_0)+4f(x_1)+2f(x_2)+...+ 2f(x_{n-1})+4f(x_{n-1})+f(x_n))$

donde $𝚫=b-a$

La regla de Simpson ofrece una mayor precisión en comparación con las sumas de Riemann para funciones suaves y bien comportadas. Es especialmente útil cuando se requiere una buena aproximación numérica de la integral y se prefiere un método más preciso que la regla del trapecio o las sumas de Riemann.

```python
import numpy as np

def funcion_a_integrar(x):
    return x**2

def regla_simpson(func, a, b, particiones):
    if particiones % 2 == 1:
        raise ValueError("El número de particiones debe ser par para la regla de Simpson.")

    delta_x = (b - a) / particiones
    x_vals = np.linspace(a, b, particiones + 1)

    suma = func(x_vals[0]) + func(x_vals[-1])
    suma += 4 * np.sum(func(x_vals[1:-1:2]))  # términos impares multiplicados por 4
    suma += 2 * np.sum(func(x_vals[2:-2:2]))  # términos pares multiplicados por 2

    resultado = suma * delta_x / 3
    return resultado

# Parámetros personalizables
limite_inferior = 0
limite_superior = 1
numero_particiones = 100

# Calcular la integral usando la regla de Simpson
resultado_simpson = regla_simpson(funcion_a_integrar, limite_inferior, limite_superior, numero_particiones)

print(f"Resultado Regla de Simpson: Área aproximada {resultado_simpson}")
```
